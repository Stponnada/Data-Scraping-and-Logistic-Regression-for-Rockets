```python
import pandas as pd
import numpy as np

df_raw = pd.read_csv('launches_raw.csv', header = [0,1,2])
#if you look at our csv file, the first 3 lines represent the header, not just the first. So when expressed as a table these 3 rows appear as one in the header.
#pd.read.csv() reads our csv file and stores it in df_raw (dataframe raw)

print("----- DataFrame Info -----")
df_raw.info()
#print some info about our data frames


print("\n--- Raw Column Names as ordered tuples --- ")
print(df_raw.columns)
# ensure that the 3 different top rows are infact arranged together as a multiindex

print("\n--- First 5 rows ---")
df_raw.head()
#print the first 5 rows of the table
```

```python
#JUPYTER NOTEBOOK CELL SUPREMACY. WE CAN NOW SIMPLY CREATE ANOTHER CELL TO WORK ON THIS DATA!!
print("\n--- First 10 rows ---")
df_raw.head(10)
#print the first 10 rows of the table
```

```python
#look at this data. In the header, it seems like the second row is actual useful information. Lets start off trying to extract just this:
new_columns = [col[1] for col in df_raw.columns]
#the first element is indexed as as element 0. Here the second element is being indesed as 1, hence col[1].

print("new, clean columns")
print(new_columns)
#show us the new clean columns

df_raw.columns = new_columns
#assign these new columns back to the original dataframe

print("\n--- First 5 rows with updated columns ---")
df_raw.head()
#print first 5 rows of new updated table
```

```python
print("\n--- First 20 rows with updated columns ---")
df_raw.head(20)
```

```python
#first, we delete all junk rows. these are rows with NaN throughout

df_working = df_raw.copy()
#create a working copy just to ensure we dont mess with the original data

import numpy as np
df_working = df_working.replace(r'^(\s|&nbsp;)*$', np.nan, regex=True)
#replace all effectively empty rows with NaN

columns_to_check = df_working.columns[1:8]
#Selects columns from index 1 to 7

df_working = df_working.dropna(subset=columns_to_check, how='all')
# in df_working drop all na rows, check columns mentioned in columns_to_check, ie 1 to 7

print("Shape after dropping all-NaN rows:", df_working.shape)
df_working.head(20)
```
