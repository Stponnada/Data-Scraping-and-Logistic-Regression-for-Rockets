```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# This creates the tool that will do the encoding for us.
le = LabelEncoder()

# We use .fit_transform() which is a convenient method that both learns the mapping
# and applies it to the 'Rocket' column in one step.
# The result is a new numerical column.
df['Rocket_ID'] = le.fit_transform(df['Rocket'])

# Display the original 'Rocket' column alongside the new 'Rocket_ID' column
print("Original and new columns side-by-side:")
print(df[['Rocket', 'Rocket_ID']].head())

#  The encoder object 'le' holds the mapping.
# We can access it using the .classes_ attribute.
print("\nHere is the key (the mapping of IDs to rocket names):")
for i, name in enumerate(le.classes_):
    print(f"ID: {i} -> Rocket: {name}")

# DataFrame 'df' now has a new column called 'Rocket_ID'
print("\nEncoding complete! Your DataFrame now has a 'Rocket_ID' column.")
```

```python
import pandas as pd
import re

# We will create a new function to clean the string for us
def clean_datetime_string(dt_str):
    # This regex pattern will find the day, month, time, and year in the string
    # It's very flexible and will work even with extra text like "(UTC) [1]-"
    match = re.search(r'(\d+)\s+([A-Za-z]+)\s+(\d{2}:\d{2}).*?(\d{4})', dt_str)
    
    if match:
        day = match.group(1)
        month = match.group(2)
        time = match.group(3)
        year = match.group(4)
        
        # Combine the found parts into a clean string for pandas to parse
        return f"{day} {month} {time} {year}"
    else:
        # If the regex doesn't find a match, return NaN
        return None

# Apply the cleaning function to the 'DateTimeYear' column
df['Cleaned_DateTime'] = df['DateTimeYear'].apply(clean_datetime_string)

# Now, we convert this much cleaner string to a proper datetime format.
# The format string '%d %B %H:%M %Y' will now work correctly!
df['DateTime_Parsed'] = pd.to_datetime(df['Cleaned_DateTime'], format='%d %B %H:%M %Y')

# --- Now, we can extract the new features! ---

# Extract the Year
df['Year'] = df['DateTime_Parsed'].dt.year

# Extract the Month (1=January, 12=December)
df['Month'] = df['DateTime_Parsed'].dt.month

# Extract the Day of the Week (0=Monday, 6=Sunday)
df['DayOfWeek'] = df['DateTime_Parsed'].dt.dayofweek

# Extract the Hour of the Day
df['Hour'] = df['DateTime_Parsed'].dt.hour

# --- Let's check the result again ---
print("New columns created from the 'DateTimeYear' column:")
print(df[['DateTimeYear', 'Year', 'Month', 'DayOfWeek', 'Hour']].head(10))
```
